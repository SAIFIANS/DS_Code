```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#strip plot
sns.stripplot(y='tip',data=tips,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
#Grouping with stripplot
sns.stripplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',hue='sex',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.violinplot(x='day', y='tip', data=tips, innner=None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip($)')
plt.show()
```
